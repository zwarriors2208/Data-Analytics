#### multiple condition
- in numpy instead of using and, or etc like we do in classic python we will use &, |, ~ etc ^

```python
df_new = df[(df['total_bill'] > 30) & (df['gender'] == 'Male')]
```

```python
df_new.head()
```

```python
df_new = df[(df['total_bill'] > 30) & ~(df['gender'] == 'Male')]
df_new.head()
```

```python
df_new = df[(df['total_bill'] > 30) & (df['gender'] != 'Male')]
```

```python
df_new.head()
```

```python
df_new = df[(df['total_bill'] >30) | (df['tip'] > 5)]
df_new.head()
```

### GroupBy
- Splitting : The data is first split into groups based on the criteria provided
- Applying : A function (or multiple functions) is applied to each group independently
- Combining: The results of the function applications

```python
df_grouped=df.groupby('gender')
print(df_grouped)
```

- - Applying
  - - min(), max(), std(), var(), count(), sum() - these functions can be also applied while applying groupBy

```python
mean_tip_by_gender=df_grouped['tip'].mean()
print(mean_tip_by_gender)
```

- - Combining

```python
df_grouped=df.groupby(['gender','day'])
print(df_grouped['tip'].mean())
```

### aggregate function __agg()__
- used to apply multiple aggregation fuvntion to differnt columns when grouping data

```python
df_grouped=df.groupby(['gender','day']).agg({'tip':['mean','sum']}).reset_index()
df_grouped
```

#### Rename the columns for clarity

```python
df_grouped.columns=['gender','day','Average Tip', 'Total Tip']
print(df_grouped)
```
